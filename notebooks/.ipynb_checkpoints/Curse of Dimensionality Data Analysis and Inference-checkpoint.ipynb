```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
np.random.seed()
```

```python
dim=(15, 3)
x = np.random.uniform(low=0, high=100, size=dim)
target = np.random.randint(low=0, high=2, size=dim[0], dtype=np.int32)
data = np.column_stack((x, target))

data
```

```python
colors = ['#f00', '0f0']
norm = matplotlib.colors.Normalize(vmin=0, vmax=1, clip=True)
mapping = cm.ScalarMappable(norm=norm, cmap=cm.winter)
```

```python
plt1 = plt.scatter(data[:, 0], np.zeros_like(data[:, 0]), c=mapping.to_rgba(data[:,3]))
plt.savefig('curse of dimensionality 1d.png')
```

```python
axis = plt.figure().gca()
axis.set_xticks(ticks=np.arange(0, 101, 10))
axis.set_yticks(ticks=np.arange(0, 101, 10))
plt.scatter(data[:, 0], data[:, 1], c=mapping.to_rgba(data[:,3]))
# plt.savefig('curse of dimensionality 2d.png')
```

```python
fig = plt.figure(figsize=(6, 6), tight_layout=False)
fig.tight_layout(pad=250.0)
ax1 = fig.add_subplot(2, 2, 1, projection='3d')
ax1.scatter(data[:, 0], data[:, 1], data[:, 2], c=mapping.to_rgba(data[:, 3]))
ax2 = fig.add_subplot(2, 2, 2)
ax2.scatter(data[:, 0], data[:, 1], c=mapping.to_rgba(data[:, 3]))
ax3 = fig.add_subplot(2, 2, 3)
ax3.scatter(data[:, 0], data[:, 2], c=mapping.to_rgba(data[:, 3]))
ax4 = fig.add_subplot(2, 2, 4)
ax4.scatter(data[:, 1], data[:, 2], c=mapping.to_rgba(data[:, 3]))
plt.savefig('curse of dimensionality 2d.png')
```
